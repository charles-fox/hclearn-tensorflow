**go.py :** The place to start! Executes a series of functions to import data, generate SURF feature vectors, cluster those down to the most useful subset, then train the RBM, before testing with a random walk through a simulated environment and outputting the figures from Saul, Prescott and Fox 2011. The code in this notebook executes this functionality. If you want to do something differenct, dig into the code or the separate python scripts listed here.

**makeMaze.py :** Generates the virtual maze environment of a specified size (not clear how flexible this is). Virtual maze includes percepts, available actions and their consequences (where you would end up) at each location in that map based on SURF extracted image features (generated through a call to SURFExtractor.py). This is essentially done by hand. Replacing this with a learned representation through experience e.g. video would be a solid goal. Calling this script spits out a lot of text to the terminal describing the SURFExtractor derived feature dictionaries.

**SURFExtractor.py :** Called by makeMaze.py. Given a folder it will extract descriptors (SURF image features here), merge them and generate the featureVectors for each image, and store them. An array of feature vectors are returned.

**paths.py :** Generates the path object, which includes information about where the agent is at any point along a path e.g. in the middle or at one end of the map, and where certain landmarks are WRT this position

**hcq.py :** Does proper wake-sleep learning as in *Fox & Prescott, Learning in a unitary coherent hippocampus*

**learnWeights.py :** Learn weights between modules of the model, based on both idealised and 'noisy' input data



**rbm.py :** Restricted Boltzmann Machine code, based on Hinton's TRBM work. Includes a number of functions for building the network, setting observed and hidden units, training weights and measuring the 'Energy'. 

  * `boltzmannProbs(W,x) :` Returns the probability of a node being on (is the weight.*x above a half?).
  
  * `trainPriorBias(hids) :` Normalises and concatenates the hidden values (I think! 30/4/14). Uses addBias(hids) to concatenate the hidden node values (their energy?). Also adds/removes tiny values to/from zeros/ones, and uses invsig to normalise the output

  * `addBias(xs) :` reshapes and concatenates input vector with a column of ones (I think! 30/4/14).  


**location.py :** Converts X Y position into 'grid cell' vector activations, based on simple boolean grid cell models of different sizes

**gui.py :** Used for plotting

**DGStateAlan.py :** Learn DG weights. 

**cffun.py :** More functions, probably from Charles Fox (hence CFfun).

  * `invsig(x) :` takes neg log of 1/x - 1. Squashes x into the range 0:1, with infinitely large outputs near those values. (Try inputing -log((1/x)-1)) into google.
